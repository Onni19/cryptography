```python
import hashlib
import math
import random
import time
```

```python
start_time = time.time()
```

```python
def XOR(final_key,padding):
  s=''
  for x in range(len(final_key)):
          if(final_key[x]==padding[x]):
              s=s+'0'
          else:
              s=s+'1'
  return s    
```

```python
def keygen(binary_key,binary_message):
    ipad='00110110'                                      #the default ipad
    opad='01011100'                                      #the default opad
    padding_ipad=ipad*int(len(binary_message)/8)         #the ipad is mutlipied until it reaches the lenght of message
    padding_opad=opad*int(len(binary_message)/8)         #the opad is mutlipied until it reaches the lenght of message
    s1=''
    s2=''
    if(len(binary_key)!=len(binary_message)):             #if the key lenght is less than message then it is padded with 0 usually 0<k<b
        no_of_zeros = (len(binary_message)-len(binary_key))*'0'
        final_key = no_of_zeros+ binary_key
    else:
        final_key=binary_key

    s1 = XOR(final_key,padding_ipad)                      #the S1 and is generated by XOR of padding of ipad and key
    s2 = XOR(final_key,padding_opad)                      #the S2 and is generated by XOR of padding of ipad and key
    return s1,s2
```

```python
def hashgenerator(message):
    hasher=hashlib.sha512()                                   #this is a pre built library that allows to generate md5 hash for a text
    sendermessage=(message).encode()                          #the message is encoded to utf-8
    hasher.update(sendermessage)
    bin_hash = "{0:08b}".format(int(hasher.hexdigest(), 16))  #the hash is generated and converted from hex to binary
    return bin_hash
```

```python
message="I am Hash"                                               #message declared by the user
binary_message = ''.join(format(ord(i), '08b') for i in message)  #message converted to binary ,0 pad with zeros, 8 pad to a total length,b use binary representation for the number
key="0079"                                                        #in the improved version the key in predefined 
binary_key = ''.join(format(ord(i), '08b') for i in key)
binary_message = ''.join(format(ord(i), '08b') for i in message)
s1,s2=keygen(binary_key, binary_message)
hash1=hashgenerator(s1+binary_message)
#print(hash1)
if(len(hash1)<len(binary_message)):                               #if the hash generated is less than b bits then it is padded
        no_of_zeros=(len(binary_message)-len(hash1))*'0'          #the paddding is done with 0's
        hash1=no_of_zeros+hash1                                   #padding is appended
hash2=hashgenerator(s2+hash1)                                     #the second round of hashing is done with S2
hash2=hex(int(hash2,2))
print("This is the final hash code :",hash2)


print ("It took ", time.time() - start_time, "to run")
```
